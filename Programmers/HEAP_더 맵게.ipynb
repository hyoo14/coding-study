```python
#2020.8.20? 100%


from heapq import heappop, heappush ,heapify, _siftup
def leftchild(p_i): return 2*p_i+1
def rightchild(p_i): return 2*(p_i+1)
def parent(i): return int((i-1)/2)
def heappush2(list01, new_elem):
    list01.append(new_elem)
    list01 = shiftup2(list01, parent(len(list01)-1),len(list01)-1, new_elem)
    return list01 
def shiftup2(list01, i, son_i, son):#좌우 자식 모두 고려
    list01[i], element = son, list01[i]
    list01[son_i] = element
    now_i = i
    while(True):
        
        #if list01[now_i] == 1:print('  ',list01)#
        
        p_i = parent(now_i)
        l_i = leftchild(p_i)
        r_i = rightchild(p_i)
        
        if l_i < len(list01) and r_i < len(list01): 
            if list01[l_i] < list01[r_i]: now_i = l_i
            else: now_i = r_i
        elif l_i < len(list01) and r_i >= len(list01): now_i = l_i
        elif l_i >= len(list01) and r_i < len(list01): now_i = r_i
        

        if list01[p_i] > list01[now_i]:
            swap1, swap2 = list01[now_i], list01[p_i]
            list01[now_i], list01[p_i] = swap2, swap1
            now_i = p_i

        else:
            #if list01[now_i] == 1:print('  ',list01[p_i],' ',list01[now_i],' ',list01)#
            break
    return list01
def shiftup(list01, i, son_i, son):
    list01[i], element = son, list01[i]
    list01[son_i] = element
    now_i = i
    while(True):
        
        #if list01[now_i] == 1:print('  ',list01)#
        
        p_i = parent(now_i)

        if list01[p_i] > list01[now_i]:
            swap1, swap2 = list01[now_i], list01[p_i]
            list01[now_i], list01[p_i] = swap2, swap1
            now_i = p_i

        else:
            #if list01[now_i] == 1:print('  ',list01[p_i],' ',list01[now_i],' ',list01)#
            break
    return list01

def heapify2 (list01):
    for i in range(len(list01)):
        #print(list01)
        element = list01[i]
        l_i, r_i = leftchild(i), rightchild(i)
        
        if l_i > len(list01)-1 and r_i > len(list01)-1:            
            #print('here1?','  ',i, '  ', element, '  ',l_i,'  ',r_i)
            break        
            
        if l_i <= len(list01)-1: 
            l_son = list01[leftchild(i)]
            if element > l_son : 
                list01 = shiftup(list01, i, l_i, l_son)
                element = list01[i]
        if r_i <= len(list01)-1: 
            r_son = list01[rightchild(i)]
            if element > r_son:
                list01 = shiftup(list01, i, r_i, r_son)    
    
    return list01
def shiftdown(list01):#좌우 자식 모두 고려
    now_i = 0
    while(True):
        
        
        p_i = now_i
        l_i = leftchild(p_i)
        r_i = rightchild(p_i)
        
        #print('p_i ',p_i,' l_i ', l_i, ' r_i ', r_i)
        
        if l_i < len(list01) and r_i < len(list01): 
            if list01[l_i] < list01[r_i]: now_i = l_i
            else: now_i = r_i
        elif l_i < len(list01) and r_i >= len(list01): now_i = l_i
        elif l_i >= len(list01) and r_i < len(list01): now_i = r_i
        else: break
        #print('     p_i', p_i,'  list01[p_i] ', list01[p_i],'  now_i ', now_i, '   list01[now_i] ',list01[now_i])

        if list01[p_i] > list01[now_i]:
            #print('            ',p_i,'  ',now_i,'  ',list01[p_i], '  ',list01[now_i])
            swap1, swap2 = list01[now_i], list01[p_i]
            list01[now_i], list01[p_i] = swap2, swap1
            #now_i = p_i

        else: break
    return list01
def heappop2(list01):
    pop_elem, swap1 = list01[0], list01[len(list01)-1]
    list01[0] = swap1
    del list01[len(list01)-1]
    #print('in ',list01,  'son이자 실질 교체용 마지막 엘리먼트 인덱스 ', len(list01)-1)
    if list01 : _siftup(list01, 0)#if list01 : list01 = shiftdown(list01)#
    #print('out ', list01)
    return pop_elem, list01

def heappop3(list01):
    
    if len(list01) == 1: return list01[0], []
    pop_elem, swap1 = list01[0], list01[len(list01)-1]
    list01[0] = swap1
    
    list01 = list01[:-1]#del list01[len(list01)-1] #list01 = list01[:-1]
    
    if list01: 
        now_i = 0
        while(True):
            p_i = now_i
            l_i = leftchild(p_i)
            r_i = rightchild(p_i)

            if l_i < len(list01) and r_i < len(list01): 
                if list01[l_i] < list01[r_i]: now_i = l_i
                else: now_i = r_i
            elif l_i < len(list01) and r_i >= len(list01): now_i = l_i
            elif l_i >= len(list01) and r_i < len(list01): now_i = r_i
            else: break
            
            if list01[p_i] > list01[now_i]:
                swap1, swap2 = list01[now_i], list01[p_i]
                list01[now_i], list01[p_i] = swap2, swap1
                
            else: break
        
    return pop_elem, list01

def heappop_0(heap):
    """Pop the smallest item off the heap, maintaining the heap invariant."""
    lastelt = heap.pop()    # raises appropriate IndexError if heap is empty
    if heap:
        returnitem = heap[0]
        heap[0] = lastelt
        _siftup_0(heap, 0)
        return returnitem
    return lastelt
def _siftup_0(heap, pos):
    endpos = len(heap)
    startpos = pos
    newitem = heap[pos]
    # Bubble up the smaller child until hitting a leaf.
    childpos = 2*pos + 1    # leftmost child position
    while childpos < endpos:
        # Set childpos to index of smaller child.
        rightpos = childpos + 1
        if rightpos < endpos and not heap[childpos] < heap[rightpos]:
            childpos = rightpos
        # Move the smaller child up.
        heap[pos] = heap[childpos]
        pos = childpos
        childpos = 2*pos + 1
    # The leaf at pos is empty now.  Put newitem there, and bubble it up
    # to its final resting place (by sifting its parents down).
    heap[pos] = newitem
    _siftdown_0(heap, startpos, pos)

def _siftdown_0(heap, startpos, pos):
    newitem = heap[pos]
    # Follow the path to the root, moving parents down until finding a place
    # newitem fits.
    while pos > startpos:
        parentpos = (pos - 1) >> 1
        parent = heap[parentpos]
        if newitem < parent:
            heap[pos] = parent
            pos = parentpos
            continue
        break
    heap[pos] = newitem

def solution(scoville, K):
    answer=0
    
    scoville = heapify2(scoville)
    #heapify(scoville)
    #print(len(scoville), end="")
    while len(scoville) >= 2:
        #heapify(scoville)#scoville = heapify(scoville)    
        #print(scoville)
        new_element = heappop(scoville)#new_element, scoville = heappop3(scoville)#
        if new_element > K: break
            
        #del scoville[0]
        #heapify(scoville)#scoville = heapify(scoville)
        new_new_element = heappop(scoville)##new_new_element, scoville = heappop3(scoville)#
        new_element += (2 * new_new_element)#new_element += (2 * heappop(scoville))
        #del scoville[0]
        scoville = heappush2(scoville,new_element)#heappush(scoville,new_element)#scoville.insert(0, new_element)
        answer += 1
    if heappop(scoville) <= K : answer = -1
    return answer



```

# 해설:

매우 힘든 여정이면서도 찜찜했던 힙문제였다는 과거의 서술이 있네..  
힙피파이 구현하고 while 문 안에서 heapify를 돌렸는데 시간초과가 떴었다고함  
그래서 heapify는 와일문 밖에서 한번만 하고, while문 안에서는 heappop과 heappush를 사용  
heappush의 경우 라이브러리 대체가 가능했지만, heappop은 대체하기 어려웠음  
heappush에서는 bottom-up방식을, heappop에서는 top-down방식을 사용하는 것이 용이  
결국 heappop은 라이브러리 코드 그대로 복붙해와도 안 돌아가서..왜 라이브러리 임포트 할 때만 효율성(시간이 적게 걸림)이 좋은 지는 모르겠  
중간에 참조한 코드는 아래와 같고 다들 구현하기보다는 임포트를 했네  
참고로 위의 heappop_0과 같이 _0이 들어간 것이 라이브러리에 있는 코드 복붙임  





# 다시 풀어 본 코드:  

```python
#acc 100 21.12.2
from heapq import heapify, heappop, heappush
def solution(scoville, K):
    answer = 0
    
    #print(scoville)
    #scoville2 = heapify(scoville)
    #print(heappop(scoville) )
    heapify(scoville)
    while True:
        
        most_ = heappop(scoville)
        if most_ >= K: break
        if len(scoville) == 0: return -1
        
        second_ = heappop(scoville)
        #print(most_,'   ',second_)
        heappush(scoville,  most_ + second_ * 2)
        #heapify(scoville)
        answer += 1
    return answer
```
