## "Join strings by multiple delimiters" prob in List

# Problem Link: https://www.geeksforgeeks.org/python-join-strings-by-multiple-delimiters/

Problem:   
list string과 delimiter string을 받음  
새로운 조합은 list string의 element 사이에 delimiter들이 들어가야함  
새로운 조합의 개수와 각 조합들을 모두 구하시오  

  
Input: list, delimiter     
Output: delimiter가 들어가는 list 조합의 개수와 각 조합들 모두 출력    

Assumptions:  
list, delimiter는 0보다 커야함    

# Solved code:

```python

from itertools import product

def sol(delimter_, list_):
    ans = []
    #delimter_, list_ = ".?", "abc"
    D, L, answer = len(delimter_), len(list_), ""
    
    if D == 0 or L ==0: return [0]

    list_num = D ** (L-1)
    d_list,l_list  = [],[]
    for i in range(D): d_list.append(delimter_[i]) #O(D)
    for i in range(L): l_list.append(list_[i]) #O(L)
    #print(d_list)

    ans.append(list_num)
    #print(list_num)
    #print("---------------------------")
    #print(product( d_list, repeat= L-1))
    for i in product( d_list, repeat= L-1): #O(D^(L-2))
        out_str = list_
        for k in range(len(i)):
            idx = 2*k+1
            out_str = out_str[:idx] + i[k] + out_str[idx:]
        #print(out_str)
        ans.append(out_str)
    return ans
```

# Test cases

```python
delimter_, list_ = ".?", "abc"
sol(delimter_, list_)
```

```python
delimter_, list_ = ".", "a"
sol(delimter_, list_)
```

```python
delimter_, list_ = "./+-=", "ab"
sol(delimter_, list_)
```

```python
delimter_, list_ = "./-+=", "abcdefg"
sol(delimter_, list_)
```

```python
delimter_, list_ = ".", "abcdefg"
sol(delimter_, list_)
```

```python
delimter_, list_ = "./", "abcdefg"
sol(delimter_, list_)
```

```python
delimter_, list_ = "", ""          #여기가 문제였나...?
sol(delimter_, list_)
```
