## "Stock Maximize" prob in DynamicProgramming

# Problem Link: https://www.hackerrank.com/challenges/stockmax/problem

Problem:   
가격들 리스트 받고  
각 가격들마다 가능한 고점에서 팔 경우의 합을 구하는 것  
  
Input: prices   
Output: 최고 수익    

Assumptions:  
  

# Solved code:

```python
#22.01.01 100% correctness100, performance100, no-timeout100, no-runtimeError100

def stockmax_sol(prices):
    
    N = len(prices)
    max_price = prices[-1]
    merit = 0
    
    for i in range(N-2, -1, -1):
        now = prices[i]
        
        if now > max_price: max_price = now
        
        merit += max_price - now
    
    return merit
```
